```python
x=np.arange(-5,5,0.2)
y1=np.sin(x)
y2=np.cos(x)
y3=np.tan(x)
fig,axs =plt.subplots(nrows=1,ncols=3,figsize=(15,3))
axs[0].plot(x,y1,'r--')
axs[1].plot(x,y2,'b--')
axs[2].plot(x,y3,'g--')
axs[0].set_title('sin(x)')
axs[1].set_title('cos(x)')
axs[2].set_title('tan(x)')
plt.show()
```

```python
# plt -> implicit figure and axes creation
x=np.linspace(-10,10,25)
plt.plot(x,x,'-',label='linear')
plt.plot(x,x**2,'--',label='quadratic')
plt.plot(x,x**3,'o',label='cubic')
plt.xlabel('my x label')
plt.ylabel('my y label')
plt.title('simple plot with implicit graph settings')
plt.legend()
plt.savefig('Plot1.png', bbox_inches='tight', dpi=300)
plt.show()
```

```python
data={
    "a": np.arange(50),
    "b": np.random.randint(0, 50, 50),
    "c": np.random.randn(50),
    "d": np.arange(50)+10 * np.random.randn(50),
}
data['d']=np.abs(data['d'])*10
fig, ax = plt.subplots(figsize=(10, 3))
ax.scatter("a", "b",c='c', s='d', data=data, cmap="tab20")
ax.set_xlabel("entry a")
ax.set_ylabel("entry b")
ax.set_title("Scatter Plot")
plt.show()
```

```python
data={
    "names": ["John", "Anna", "Peter", "Linda"],
    "english": [95, 85, 98, 90],
    "math": [90, 95, 92, 85],
    "hindi": [98, 90, 95, 60],
}
```

```python
def my_custom_plotter(ax,x_values,y_values,param_dict):
    out = ax.bar(x_values,y_values,**param_dict)
    ax.set_title("Visualizing marks")
    ax.set_xlabel("Students")
    ax.set_ylabel("Marks")
    return out
```

```python
fig, axs=plt.subplots(1,3,figsize=(15,3))
my_custom_plotter(axs[0], 'names','english', {'data':data,'color':'red','alpha':.3})
my_custom_plotter(axs[1], 'names','math', {'data':data,'color':'green','alpha':.6})
my_custom_plotter(axs[2], 'names','hindi',{'data':data, 'color':'blue'})
```
