- Categorial graphs
    - columns that have text are generally categorial
    - the first step is to count the numbers of uique values in the column
    - then plot the count of each category
    - following charts are a good way to visualize categorial data
        - bar chart
        - pie chart
        - donut chart
        - line chart(when too many categories)
        - area chart(when too many categories)

```python
survival_count= titanic['alive'].value_counts()
survival_count
```

```python
survival_count.plot(kind='pie',autopct='%.1f%%',title='Survival Rate')
```

```python
survival_count.plot(
    kind='pie',
    autopct='%1.1f%%',
    title='Survival Rate',
    labels=['Survived', 'Not Survived'],
    colors=['#66b3ff','#ff9999'],
    labeldistance=1.1,
    wedgeprops = {
        'width':0.6,
        'edgecolor':'red',
        'linewidth':1.5,
        'linestyle':'dashdot',
})
```

```python
dots=sns.load_dataset('dots')
sns.relplot(
    data=dots,
    x='time',
    y='firing_rate',
    hue='choice',
    row='align',
    col='choice',
    kind='line',
    size='coherence',
    style='choice',
)
```
