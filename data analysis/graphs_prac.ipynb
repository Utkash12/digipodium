```python
iris=sns.load_dataset('iris') #about iris flower
iris.petallength.plot.hist()
```

```python
plt.hist('sepal_length', data=iris, bins=20)
plt.show()
```

```python
fig,axs=plt.subplots(2,2,figsize=(10,6), layout='constrained')
axs[0][0].hist('petal_width', data=iris, bins=25, color='red', label='petal_width')
axs[0][0].legend()
axs[0][0].set_title('petal_width')



axs[0][1].hist('petal_length', data=iris, bins=25, color='blue', label='petal_length')
axs[0][1].legend()
axs[0][1].set_title('petal_length')



axs[1][0].hist('sepal_length', data=iris, bins=20, color='green', label='sepal_length')
axs[1][0].legend()
axs[1][0].set_title('sepal_length')



axs[1][1].hist('sepal_width', data=iris, bins=20, color='yellow', label='sepal_width')
axs[1][1].legend()
axs[1][1].set_title('sepal_width')
plt.show()
```

```python
fig, axs=plt.subplots(2,2,figsize=(10,6))
hindi=np.array([40,42,44,46,48,50,40])
english=np.array([60,55,76,44,55,66])
maths=np.array([65,67,76,65,44,55])
physics=np.array([77,75,45,65,73,44,65])
axs[0][0].hist(hindi, color='red', align='mid',label='Hindi')
axs[0][0].legend()
axs[0][0].set_title('Hindi')
axs[0][1].hist(english, color='green', align='mid',label='English')
axs[0][1].legend()
axs[0][1].set_title('English')
axs[1][0].hist(maths, color='blue', align='mid',label='Maths')
axs[1][0].legend()
axs[1][0].set_title('Maths')
axs[1][1].hist(physics, color='yellow', align='mid',label='Physics')
axs[1][1].legend()
axs[1][1].set_title('Physics')
plt.show()
```
