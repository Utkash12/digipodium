```python
plt.style.use('seaborn-whitegrid')
```

```python
fig,axs=plt.subplots(2,2,figsize=(10,10), layout='constrained')
axs[0][0].hist('petal_width', data=iris, bins=25, color='red', label='petal_width')
axs[0][0].legend()
axs[0][0].set_title('petal_width')



axs[0][1].hist('petal_length', data=iris, bins=25, color='blue', label='petal_length')
axs[0][1].legend()
axs[0][1].set_title('petal_length')



axs[1][0].hist('sepal_length', data=iris, bins=20, color='green', label='sepal_length')
axs[1][0].legend()
axs[1][0].set_title('sepal_length')



axs[1][1].hist('sepal_width', data=iris, bins=20, color='yellow', label='sepal_width')
axs[1][1].legend()
axs[1][1].set_title('sepal_width')
plt.show()
```
