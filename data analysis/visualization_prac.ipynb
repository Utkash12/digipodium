```python
hindi=np.array([60,70,73,60,53,56])
maths=np.array([40,50,50,30,60,70])
english=np.array([90,70,50,60,80,45])
plt.xlabel('students',fontsize=10,fontweight='bold',color='red',fontstyle='italic',labelpad=10,rotation=25)
plt.ylabel('marks',fontsize=10,fontweight='bold',color='red',fontstyle='italic',labelpad=10,rotation=25)
plt.title('marks of students',fontsize=10,fontweight='bold',color='red',fontstyle='italic',pad=10,rotation=15)
plt.plot(hindi,'r--',marker='*')
plt.plot(english,'g--',marker='o')
plt.plot(maths,'b--',marker='^')
plt.show()
```

```python
x=np.arange(-5,5,0.2)
y1=np.sin(x)
y2=np.cos(x)
y3=np.tan(x)
fig,axs =plt.subplots(nrows=3,ncols=1,figsize=(5,10))
axs[0].plot(x,y1,'r--')
axs[1].plot(x,y2,'b--')
axs[2].plot(x,y3,'g--')
axs[0].set_title('sin(x)')
axs[1].set_title('cos(x)')
axs[2].set_title('tan(x)')
plt.show()
```

```python
# plt -> implicit figure and axes creation
x=np.linspace(-10,10,25)
plt.plot(x,x,'-',label='one')
plt.plot(x,x**2,'--',label='two')
plt.plot(x,x**3,'o',label='three')
plt.plot(x,x**4,'*',label='four')
plt.xlabel('my x label')
plt.ylabel('my y label')
plt.legend()
plt.title('simple plot with implicit graph settings')
plt.savefig('Plot1.png', bbox_inches='tight', dpi=300)
plt.show()
```

```python
hindi=np.array([60,70,40,50,70,55,60])
english=np.array([90,60,70,80,50,70,60])
maths=np.array([60,70,80,50,40,80,70])
plt.plot(hindi,'b^-', label='hindi')
plt.plot(english, 'g*-', label='english')
plt.plot(maths, 'ro-', label='maths')
plt.legend()
plt.title('marks of students')
plt.xlabel('students')
plt.ylabel('marks')
plt.grid()
plt.show()
```

```python
hindi=np.array([60,70,40,50,70,55,60])
english=np.array([90,60,70,80,50,70,60])
maths=np.array([60,70,80,50,40,80,70])
physics=np.array([60,50,70,80,55,75])
plt.subplot(2,2,1)
plt.plot(hindi,'b^-', label='hindi')
plt.grid()
plt.legend()
plt.subplot(2,2,2)
plt.plot(english, 'g*-', label='english')
plt.legend()
plt.grid()
plt.subplot(2,2,3)
plt.legend()
plt.plot(maths, 'ro-', label='maths')
plt.grid()
plt.legend()
plt.subplot(2,2,4)
plt.plot(physics, 'y*-', label='physics')
plt.legend()
plt.grid()
plt.show()
```
