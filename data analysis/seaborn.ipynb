```python
tips=sns.load_dataset('tips')
print(tips)
```

```python
print(tips.head())
```

```python
sns.relplot(
    data=tips,
    x="total_bill", y="tip", row="size",
    hue="smoker", style="smoker", size="size",
)
```

```python
sns.relplot(x="total_bill", y="tip", hue="smoker", data=tips)
```

```python
iris=sns.load_dataset('iris')
print(iris)
```

```python
sns.set_theme()
sns.relplot(x="sepal_length",y="sepal_width",hue="petal_length",col="species", data=iris);
```
